```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def bifDiag(x0, range_,a ,b):
  x1 = x0
  p = []
  X = []
  par_range  = np.linspace(a, b, 100)
  for par in par_range:
    x = x1
    for i in range(range_):
      x = par*x*(1-x*x)
      if(i > 30):
        p.append(par)
        X.append(x)
  return p, X
```

```python
fig, ax = plt.subplots(figsize = (18, 13))
r = np.linspace(-1, 1, 20)
for i in r:
    p, X = bifDiag(i*0.0001, 300, -3, 3)
    ax.plot(p, X, 'b.',  markersize=0.5)
    ax.scatter(p, X, c = 'blue', s = 0.1)
plt.show()
```
